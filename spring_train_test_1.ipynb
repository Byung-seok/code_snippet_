```python
target = SAE9254_dataset['NTM입측온도']
dataset = SAE9254_dataset.drop(['NTM입측온도'], axis = 1)
```

```python
train_dataset = dataset.iloc[:test_start_idx]
train_target = target.iloc[:test_start_idx]
test_dataset = dataset.iloc[test_start_idx:]
test_target = target.iloc[test_start_idx:]
```

```python
train_stats = train_dataset.describe()
train_stats = train_stats.transpose()

def norm(x):
    return (x - train_stats['mean']) / train_stats['std']

train_norm = norm(train_dataset)
test_norm = norm(test_dataset)
```

```python
train_norm = train_norm.astype('float32').to_numpy()
train_target = train_target.astype('float32').to_numpy()

test_norm = test_norm.astype('float32').to_numpy()
test_target = test_target.astype('float32').to_numpy()
```

```python
num_epochs = 300
size_batch = 4
rate_learning = 1e-3

config = {
    'str_model': 'mlp',
    'str_dataset': 'Waterbox',
    'num_epochs': num_epochs,
    'rate_learning': rate_learning,
    'size_batch': size_batch,
}
```

```python
wandb.init(project="waterbox_mlp", entity = "jbee22", config = config)
```

```python
dataset_train = tf.data.Dataset.from_tensor_slices((train_norm, train_target))
dataset_train = dataset_train.batch(size_batch)

dataset_test = tf.data.Dataset.from_tensor_slices((test_norm, test_target))
dataset_test = dataset_test.batch(size_batch)
```

```python
class MLP(tf.keras.Model):
    def __init__(self):
        super().__init__()

        self.layer_fc_1 = tf.keras.layers.Dense(128, activation=tf.nn.relu)
        self.layer_fc_2 = tf.keras.layers.Dense(32, activation=tf.nn.relu)
        self.layer_fc_3 = tf.keras.layers.Dense(16, activation=tf.nn.relu)
        self.layer_fc_4 = tf.keras.layers.Dense(1)

    def call(self, inputs):
        outputs = inputs

        outputs = self.layer_fc_1(outputs)
        outputs = self.layer_fc_2(outputs)
        outputs = self.layer_fc_3(outputs)
        outputs = self.layer_fc_4(outputs)

        return outputs
```

```python
model = MLP()
optimizer = tf.keras.optimizers.Adam(learning_rate=rate_learning)
loss = tf.keras.losses.mean_squared_error

metric_mse_train = tf.keras.metrics.MeanSquaredError()
metric_mse_test = tf.keras.metrics.MeanSquaredError()
metric_train = tf.keras.metrics.RootMeanSquaredError()
metric_test = tf.keras.metrics.RootMeanSquaredError()

model.build((None, train_norm.shape[1]))
model.summary()
```

```python
@tf.function
def step_train(X, by):
    with tf.GradientTape() as tape:
        preds_ = model(X)
        loss_ = loss(by, preds_)

    grads_ = tape.gradient(loss_, model.trainable_weights)
    optimizer.apply_gradients(zip(grads_, model.trainable_weights))

    metric_mse_train.update_state(by, preds_)
    metric_train.update_state(by, preds_)

@tf.function
def step_test(X, by):
    preds_ = model(X)
    loss_ = loss(by, preds_)

    metric_mse_test.update_state(by, preds_)
    metric_test.update_state(by, preds_)
```

```python
for ind_epoch in range(0, num_epochs):
    for X_batch, by_batch in dataset_train:
        step_train(X_batch, by_batch)

    for X_batch, by_batch in dataset_test:
        step_test(X_batch, by_batch)

    loss_train = metric_mse_train.result()
    acc_train = metric_train.result()
    loss_test = metric_mse_test.result()
    acc_test = metric_test.result()

    metric_mse_train.reset_states()
    metric_train.reset_states()
    metric_mse_test.reset_states()
    metric_test.reset_states()

    print(f'{ind_epoch + 1} EPOCH: loss_train {loss_train:.4f} rmse_train {acc_train:.4f}')
    print(f'loss_test {loss_test:.4f} rmse_test {acc_test:.4f}')
    
    wandb.log({
        'ind_epoch': ind_epoch + 1,
        'loss_train': loss_train,
        'rmse_train': acc_train,
        'loss_test': loss_test,
        'rmse_test': acc_test,
    })
```

```python
def build_network(input_shape):
    model = tf.keras.models.Sequential()
    model.add(tf.keras.layers.Dense(128, activation = 'relu', input_shape = input_shape))
    model.add(tf.keras.layers.Dense(32, activation = 'relu'))
    model.add(tf.keras.layers.Dense(16, activation = 'relu'))
    model.add(tf.keras.layers.Dense(1))
    model.compile(optimizer = tf.keras.optimizers.Adam(learning_rate=rate_learning), loss = 'mse', metrics = [tf.keras.metrics.RootMeanSquaredError()])
    return model
```

```python
model = build_network((train_norm.shape[1],))
model.summary()
```

```python
model.fit(train_norm, train_target, epochs = num_epochs, batch_size = size_batch)
```
